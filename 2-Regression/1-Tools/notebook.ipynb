```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

Load the diabetes dataset

```python
x,y=datasets.load_diabetes(return_X_y=True)
```

Get the names of the columns

```python
feature_names = datasets.load_diabetes().feature_names
import pandas as pd
df = pd.DataFrame(x, columns=feature_names)
df['target'] = y
print(df.head())
```

Print the shape of the data and the first row

```python
print(x.shape)
```

```python
print (x[0])
```

Extract only bmi data (column 2)

```python
X=x[:,2]
print (X.shape)
```

Reshape o a 2D array to train 

```python
X=X.reshape((-1,1))
print(X.shape)
```

Split data into training and testing

```python
X_train, X_test, y_train, y_test=model_selection.train_test_split(X,y, test_size=0.33)
```

Create a Linear regression model and train it with our data

```python
model=linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Predict using our test data

```python
y_pred=model.predict(X_test)
```

```python
#Create a scatter plot
plt.scatter(X_test, y_test, color='black')
#Plot the predictions
plt.plot(X_test, y_pred, color='blue', linewidth=3)
#Add labels and a title
plt.xlabel('Scaled BMIs')
plt.ylabel("Disease Progression")
plt.title('A graph plot showing diabetes progression against BMI')
#Draw the plot
plt.show
```
